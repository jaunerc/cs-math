# Calculus 1 - Linear Functions

```python
import numpy as np
from matplotlib import pyplot as plt
```

Linear functions are usually the first function family for students. A function is more generally a **relation between two sets**. Each element of set $X$ maps to exactly one element in set $Y$. Maybe you will face the notation $f: X \rightarrow Y$. That is a way to express _elements of set $X$ maps to elements of set $Y$ over the function $f$_.

Linear functions have a **constant slope** and the graph of a linear function is just a straight line. One way to write linear functions is $y = f(x) = mx + b$. Where $m$ is the slope and $b$ is the y-intercept.

![linear function](../images/linear-function.png)

Let's plot the graph of the function $f(x) = 2x + 5$ in the next section as an example (the plot is just over a finite interval).

```python
x = np.linspace(-1, 10) # evenly spaced values in the interval [-1, 10]
y = 2*x + 5

plt.figure(figsize=(8,8))
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y)
```

## Typical exercises

### 1 - Find $f(x)$ with the slope $m$ so that the point $P$ lie on its graph
The slope $m$ is given therefore only $b$ is missing

$$f(x) = mx + b$$

When the point $P$ lies on the graph

$$P = (x_p,y_p)$$

the following equation must fullfill

$$f(x_p) = y_p = mx_p + b$$
$$b = y_p - mx_p$$

```python
P = (3, 3)
m = 2
b = P[1] - m * P[0]

x = np.linspace(-5,5)
y = m * x + b

plt.figure(figsize=(8,8))
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y, 'b-', label='function graph')
plt.plot(P[0], P[1], 'ro', label='point')
plt.legend(loc="upper left")
```

### 2 - Find $f(x)$ so that two given points $P_1$ and $P_2$ lie on its graph
Now with two points

$$ P_1 = (x_{p1},y_{p1}), P_2 = (x_{p2},y_{p2}) $$

The slope can be written as

$$ m = \frac{\Delta y}{\Delta x} = \frac{y_{p2} - y_{p1}}{x_{p2} - x_{p1}} $$

Then the same procedure as the exercise 1 with one point

$$f(x_{p1}) = y_{p1} = mx_{p1} + b$$
$$b = y_{p1} - mx_{p1}$$

```python
P1 = (2, 3)
P2 = (4, 1)

m = (P2[1] - P1[1]) / (P2[0] - P1[0])
b = P1[1] - m * P1[0]

x = np.linspace(-5,5)
y = m * x + b

plt.figure(figsize=(8,8))
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y, 'b-', label='function graph')
plt.plot(P1[0], P1[1], 'ro', label='points')
plt.plot(P2[0], P2[1], 'ro')
plt.legend(loc="upper left")
```

### 3 - Find the point of intersection of two linear functions
First of all there are three possibilities for this kind of exercise. Either there is exactly one point of intersection or no intersection or an infinite number of intersections.

1. Exactly **one** intersection if the slope of two functions is different
2. **No** intersections if the slope of two functions is equal but the y-intercept is different
3. **$\infty$** intersections if the slope and the y-intercept of two functions are equals

Lets say there are two functions $f_1(x) = m_1x + b_1$ and $f_2(x) = m_2x + b_2$

If there's an intersection there must be one $x_s$ so that.

\begin{align}
f_1(x) &= f_2(x) \\
m_1x_s + b_1 &= m_2x_s + b_2 \\
m_1x_s - m_2x_s &= b_2 - b_1 \\
x_s(m_1 - m_2) &= b_2 - b_1 \\
x_s &= \frac{b_2 - b_1}{m_1 - m_2} \\
\end{align}

Then
$$f_1(x_s) = f_2(x_s) = y_s$$

The point of intersection is
$$P(x_s, y_s)$$

```python
x = np.linspace(-5, 5)

y1 = 2 * x - 2
y2 = -3 * x + 3

y3 = 2 * x + 2
y4 = 2 * x

y5 = -2 * x + 2
y6 = -2 * x + 2

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,8))

ax1.set_title('1 point of intersection')
ax1.plot(x, y1)
ax1.plot(x, y2)

ax2.set_title('No intersection')
ax2.plot(x, y3)
ax2.plot(x, y4)

ax3.set_title('intinite number of intersections')
ax3.plot(x, y5)
ax3.plot(x, y6)
```
